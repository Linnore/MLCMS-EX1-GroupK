```python
from matplotlib import pyplot as plt
```

```python
with open(r"../output/Test7_results.txt", "r") as input:
    pdlines = input.readlines()[1:]

age_tspeed = {} # assigned speed 
age_sspeed = {} # simulated speed
for line in pdlines:
    pdinfo = list(map(float, line.split("\t")))
    pdinfo[1] = int(pdinfo[1])
    if age_tspeed.get(pdinfo[1], -1) == -1:
        age_tspeed[pdinfo[1]] = [pdinfo[2]]
        age_sspeed[pdinfo[1]] = [pdinfo[3]]
    else:
        age_tspeed[pdinfo[1]].append(pdinfo[2])
        age_sspeed[pdinfo[1]].append(pdinfo[3])

age_max_speed = {}
age_min_speed = {}
age_mean_speed = {}
for key in age_sspeed.keys():
    age_max_speed[key] = max(age_sspeed[key])
    age_min_speed[key] = min(age_sspeed[key])
    age_mean_speed[key] = sum(age_sspeed[key])/len(age_sspeed[key])
```

```python
age = sorted(list(age_mean_speed.keys()))
mean_speed = []
max_speed = []
min_speed = []
for i in age:
    mean_speed.append(age_mean_speed[i])
    max_speed.append(age_max_speed[i])
    min_speed.append(age_min_speed[i])
plt.plot(age, mean_speed, age, max_speed, age, min_speed)
```

```python
max_speed = []
for i in age:
    max_speed.append(age_max_speed[i])
plt.plot(age, max_speed)
```
